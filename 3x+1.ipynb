```python
def even(k):
    b=k/2
    return(int(b))
def odd(k):
    b=3*k+1
    return(b)
```

```python
import time 
import pandas as pd 
import matplotlib as plt
t=[]
for h in range(1,100):
    ls=[]
    s=h
    di={}
    while True:
#         print(s)
        if s%2==0:
            s=even(s)
            ls.append(s)
            if s==1:
                break
        else:
            s=odd(s)
            ls.append(s)
            if s==1:
                break
#         time.sleep(1)
    di["h"]=h
    for v in range(len(ls)):
        di[str(v)]=ls[v]
    t.append(di)
    ls=[]
    di={}
df=pd.DataFrame(t)
df.head()
df.plot.line()
plt.gca().axes.get_xaxis().set_visible(False)
```
